```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read in the data gleaned from Google Earth images (see readme.md)
#  df = pd.read_csv('GE_groin_data.csv', parse_dates=['Date']) # version with only Google Earth images
df = pd.read_csv('Sandwich_GE_Groin_Analysis_v2.csv', parse_dates=['Date'])
```

```python
df
```

```python
mn = df.mean()
sd = df.std()
print(mn)
print(sd)
```

```python
fig, ax = plt.subplots()
df.mean().plot(kind='line',linewidth=4,yerr=df.std(),color="#404040",\
               figsize=(10,3),ax=ax,grid=False,ylim=(-2,2),xlim=(-.5,10.5))

plt.plot([-.5, 10.5],[0, 0],'--k')
plt.show()
```

```python
for i in range(1,12):
    df.iloc[:,[i]].plot.hist(density=True, bins=[-2.5, -1.5, -.5,.5, 1.5, 2.5 ],\
                             ylim=[0,1],xlim=[-2.5,2.5],facecolor='#404040', align='mid', rwidth=1,\
                             figsize=(3.,3.)) #legend=("")
    plt.yticks([])
    plt.xticks([])
    plt.ylabel("")
    plt.plot([0, 0],[0, 1],'--k')
    figname = 'histl'+str(i)
    plt.savefig(figname,dpi=200)
```

```python
fig, ax = plt.subplots()
for i in range(1,12):
    df.mean().plot(kind='bar',yerr=df.std(),colormap='OrRd_r',edgecolor='black',grid=False,\
               figsize=(8,2),ax=ax,position=0.45,error_kw=dict(ecolor='black',elinewidth=3),width=0.8)
    plt.show()
```
