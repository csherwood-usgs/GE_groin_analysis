```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read in the data gleaned from Google Earth images (see readme.md)
df = pd.read_csv('GE_groin_data.csv', parse_dates=['Date'])
```

```python
df
```

```python
mn = df.mean()
sd = df.std()
print(mn)
print(sd)
```

```python
fig, ax = plt.subplots()
df.mean().plot(kind='line',linewidth=3,yerr=df.std(),color='black',\
               figsize=(10,3),ax=ax,grid=True)
plt.show()
```

```python
fig, ax = plt.subplots()
df.mean().plot(kind='bar',yerr=df.std(),colormap='OrRd_r',edgecolor='black',grid=False,\
               figsize=(8,2),ax=ax,position=0.45,error_kw=dict(ecolor='black',elinewidth=3),width=0.8)
plt.show()
```

```python
df.set_index('Date').plot()
plt.legend(loc='upper right',bbox_to_anchor=(1.5,1.2))
```
